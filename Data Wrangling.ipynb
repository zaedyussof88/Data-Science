```python
#Import necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Export dataset

URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
data = pd.read_csv(URL, names = headers)
```

```python
#Identify and handle missing values
#Replace '?' to 'NaN'

data.replace("?", np.nan, inplace = True)
data.head()
```

```python
#Evaluate missing data

missing_data = data.isnull()
missing_data.head()
```

```python
#Count missing values in each column

for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")    
```

Based on the summary above, each column has 205 rows of data, seven columns containing missing data:

1. "normalized-losses": 41 missing data
2. "num-of-doors": 2 missing data
3. "bore": 4 missing data
4. "stroke" : 4 missing data
5. "horsepower": 2 missing data
6. "peak-rpm": 2 missing data
7. "price": 4 missing data

```python
#Calculate the average of the column

avg_norm_loss = data["normalized-losses"].astype("float").mean(axis = 0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
#Replace "NaN" by mean value in "normalized-losses" column

data['normalized-losses'] = data['normalized-losses'].fillna(0)
```

```python
#Calculate the mean value for 'bore' column

avg_bore = data['bore'].astype('float').mean(axis = 0)
print("Average of bore:", avg_bore)
```

```python
#Replace "NaN" by mean value

data["bore"].replace(np.nan, avg_bore, inplace = True)
data
```

```python
#Calculate the mean value for the 'stroke' column

avg_stroke = data['stroke'].astype('float').mean(axis = 0)
print("Average of stroke:", avg_stroke)
```

```python
#Replace "NaN" by mean value

data["stroke"].replace(np.nan, avg_stroke, inplace = True)
data
```

```python
#Calculate the mean value for the 'horsepower' column

avg_horsepower = data['horsepower'].astype('float').mean(axis = 0)
print("Average of horsepower:", avg_horsepower)
```

```python
#Replace "NaN" by mean value

data["horsepower"].replace(np.nan, avg_horsepower, inplace = True)
data
```

```python
#Calculate the mean value for 'peak-rpm' column

avg_peakrpm = data['peak-rpm'].astype('float').mean(axis = 0)
print("Average of peak-rpm:", avg_peakrpm)
```

```python
#Replace "NaN" by mean value

data["peak-rpm"].replace(np.nan, avg_peakrpm, inplace = True)
data
```

```python
#See which values are present in a particular column

data['num-of-doors'].value_counts()
```

```python
#Common type for 'num-of-doors'

data['num-of-doors'].value_counts().idxmax()
```

```python
#replace the missing 'num-of-doors' values by the most frequent

data['num-of-doors'].value_counts()
data['num-of-doors'].value_counts().idxmax()
data["num-of-doors"].replace(np.nan, "four", inplace = True)
```

```python
#Drop whole row with NaN in "price" column

data.dropna(subset = ["price"], axis = 0, inplace = True)

#Reset index, because we droped two rows

data.reset_index(drop = True, inplace = True)
```

```python
#Convert data types to proper format

data[["bore", "stroke"]] = data[["bore", "stroke"]].astype("float")
data[["normalized-losses"]] = data[["normalized-losses"]].astype("int")
data[["price"]] = data[["price"]].astype("float")
data[["peak-rpm"]] = data[["peak-rpm"]].astype("float")
```

```python
#List the columns after the conversion

data.dtypes
```

```python
data.describe()
```

```python
#Data Standardization

data.head()
```

```python
#Convert mpg to L/100km by mathematical operation (235 divided by mpg)

data['city-L/100km'] = 235/data["city-mpg"]
data.head()
```

```python
#Convert mpg to L/100km by mathematical operation (235 divided by mpg)

data['highway-L/100km'] = 235/data["highway-mpg"] 

#Rename column name from "highway-mpg" to "highway-L/100km"

data.rename(columns = {'"highway-mpg"':'highway-L/100km'}, inplace = True)
data.head()
```

```python
#Data Normalization
#Replace (original value) by (original value)/(maximum value)

data['length'] = data['length']/data['length'].max()
data['width'] = data['width']/data['width'].max()
```

```python
#Normalize the column 'height'

data['height'] = data['height']/data['height'].max()

#Show the scaled columns

data[["length","width","height"]].head()
```

```python
#Binning Data In Pandas
#Convert data to correct format

data["horsepower"] = data["horsepower"].astype(int, copy = True)
```

```python
bins = np.linspace(min(data["horsepower"]), max(data["horsepower"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
data['horsepower-binned'] = pd.cut(data['horsepower'], bins, labels = group_names, include_lowest = True )
data[['horsepower','horsepower-binned']].head(20)
```

```python
data["horsepower-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
```

```python
#Plot the distribution of each bin.

pyplot.bar(group_names, data["horsepower-binned"].value_counts())

plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#Draw historgram of attribute "horsepower" with bins = 3

plt.pyplot.hist(data["horsepower"], bins = 3)

#Set x/y labels and plot title

plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
