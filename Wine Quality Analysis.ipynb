```python
#Import all relevent libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Import dataset

data = pd.read_csv('winequality-redR2.csv')
```

```python
#Show 10 of the head data

data.head(10)
```

```python
#Show 5 of the tail data

data.tail(5)
```

```python
#Find the number of column, observation, data types of each variables,

data.info()
```

```python
#find the min, max, average and standard deviation for variables sulphates & free sulfur dioxide

print("Min: ")
print(data[['sulphates', 'free sulfur dioxide']].min())
print()
print("Max: ")
print(data[['sulphates', 'free sulfur dioxide']].max())
print()
print("Average: ")
print(data[['sulphates', 'free sulfur dioxide']].mean())
print()
print("Standard Deviation: ")
print(data[['sulphates', 'free sulfur dioxide']].std())
```

```python
#Calculate mean fixed acidity & citric acid for each volatile acidity rank

data.groupby('volatile acidity')[['fixed acidity', 'citric acid']].mean()
```

```python
#to show subset of rows in which chlorides >= 0.1 and residual sugar == 1.6

subsetdata = data[(data['chlorides'] >= 0.1) & (data['residual sugar'] == 1.6)]
print(subsetdata)
```

```python
#to plot a histogram with 10 bins. Figure size is 20, 20

data_plot = data['quality']

plt.figure(figsize = (20, 20))
plt.hist(data_plot, bins = 10, edgecolor = 'black')
plt.title('Quality of Wine')
plt.xlabel('Quality')
plt.ylabel('Frequency')

plt.show()
```

```python
#Create a heatmap using seaborn library
#import seaborn library
#correlation matrix

import seaborn as sns
corr_matrix = data.corr()

print("Correlation Matrix:")
print(corr_matrix)

fig, ax = plt.subplots(figsize = (10,10))
sns.heatmap(corr_matrix, annot = True,ax = ax)
plt.show()
```

```python
#to plot a scatter plot for volatile acidity and quality

volatile_acidity = data['volatile acidity']
quality = data['quality']

plt.figure(figsize = (10, 6))
plt.scatter(volatile_acidity, quality, alpha = 0.5, color = 'blue')
plt.title('Volatile Acidity vs Quality')
plt.xlabel('Volatile Acidity')
plt.ylabel('Quality')

plt.show()
```
